## **Export SWOT data to the S3 Endpoint** 

```python
%%time

##################################
#Imports

import s3fs
import numpy as np
import time
import os
```

```python
# Define local directoryand s3 point
local_dir = "/Users/contrema/Documents/Postdoc/SWOTv2/L3/test/"

s3_bucket = "project-meom-ige/OMIP/"
new_s3_dir ="SWOT_L3"  # <-- this will be the new folder in the bucket
```

```python

# Load S3
local_dir_name = os.path.basename(local_dir.rstrip("/"))
fs = s3fs.S3FileSystem( anon=True, endpoint_url="https://minio.lab.dive.edito.eu", use_ssl=False ) 

for root, dirs, files in os.walk(local_dir):
    for file in files:
        if file.endswith(".nc"):  # only upload NetCDF files
            local_path = os.path.join(root, file)
            #print(local_path)
            # Compute S3 path relative to local_dir
            relative_path = os.path.relpath(local_path, local_dir)
            #print(relative_path)
            s3_path = os.path.join(new_s3_dir, local_dir_name, relative_path).replace("\\", "/")
            #print(s3_path)
            # Upload
            fs.put(local_path, "s3://"+s3_bucket+s3_path)
            print(f"Uploaded {local_path} to s3://{s3_bucket}/{s3_path}")
```
